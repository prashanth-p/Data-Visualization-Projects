# Molecular Motion:

```python

import matplotlib.pyplot as plt

from random import choice

class MolecularMotion():
    def __init__(self,num_points):
        self.num_points = num_points
        
        #assign initial value at (0,0)
        self.xval = [0]
        self.yval = [0]
        
    def randomMolecularMotion(self):
        while(len(self.xval) < self.num_points):

            xdir = choice([-1,1])
            xdist = choice(list(range(5)))
            xstep = xdir * xdist
                        
            ydir = choice([-1,1])
            ydist = choice(list(range(5)))
            ystep = ydir * ydist
            
            if (xstep==0 and ystep==0):
                continue
                
            x_nextStep = self.xval[-1] + xstep
            y_nextStep = self.yval[-1] + ystep
            
            self.xval.append(x_nextStep)
            self.yval.append(y_nextStep)
            
        
```

# Display using matplotlib

```python
def main():
    molecule = MolecularMotion(500)
    molecule.randomMolecularMotion()
    
    plt.plot(molecule.xval, molecule.yval,linewidth=1)
    #Emphasize the first point and the last point
    plt.scatter(0,0, c='red',edgecolor='none',s=100)
    plt.scatter(molecule.xval[-1],molecule.yval[-1],c='green', edgecolor='none', s=100)
    plt.show()
    
if __name__ == '__main__':
    main()
```

# Display using Pygal

```python
import pygal

molecular_motion = pygal.XY()

molecule = MolecularMotion(500)
molecule.randomMolecularMotion()

# molecular_motion.x_labels = molecule.xval
molecular_motion.add("Molecular Motion", [(k,v) for k,v in zip(molecule.xval,molecule.yval)])

molecular_motion.title = "Random Molecular Motion using Pygal"
molecular_motion.x_title = "X direction distance"
molecular_motion.y_title = "Y direction distance" 
molecular_motion.render_to_file('images/pygal_motion.svg')
```

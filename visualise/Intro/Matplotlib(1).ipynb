# x vs sqr(x) using matplotlib

```python
# Test matplotlib
import matplotlib.pyplot as plt

squares = [1, 4, 9, 16, 25]
plt.plot(squares, linewidth=5)

# Set chart title and label axis
plt.title("Square Numbers", fontsize=24)
plt.xlabel("value", fontsize=18)
plt.ylabel("Square of x", fontsize=18)

# Set size of tick labels
plt.tick_params(axis='both', labelsize=14)

print(plt.show())
```

```python
# Test matplotlib
import matplotlib.pyplot as plt

input_values = [1,2,3,4,5]
squares = [1, 4, 9, 16, 25]
plt.plot(input_values,squares, linewidth=5)

# Set chart title and label axis
plt.title("Square Numbers", fontsize=24)
plt.xlabel("value", fontsize=18)
plt.ylabel("Square of x", fontsize=18)

# Set size of tick labels
plt.tick_params(axis='both', labelsize=14)

print(plt.show())
```

# Scatter Plot

```python
import matplotlib.pyplot as plt

plt.scatter(2,4,s=200)

#Set chart title and label axis
plt.title("Square Numbers", fontsize=24)
plt.xlabel("value", fontsize=14)
plt.ylabel("square", fontsize=14)

#Set size of tick labels
plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```

```python
# Test matplotlib
import matplotlib.pyplot as plt

xvalues = list(range(1,251))
yvalues = [x**2 for x in xvalues]

plt.scatter(xvalues,yvalues,c = 'red', edgecolor='none',s=50)

# Set chart title and label axis
plt.title("Scatter Plot Square Numbers", fontsize=24)
plt.xlabel("value", fontsize=18)
plt.ylabel("Square of x", fontsize=18)

# Set size of tick labels
plt.tick_params(axis='both',which='major', labelsize=14)

plt.show()
```

# Colormap

```python
# Test matplotlib
import matplotlib.pyplot as plt

xvalues = list(range(1,251))
yvalues = [x**2 for x in xvalues]

plt.scatter(xvalues,yvalues,c=yvalues,cmap=plt.cm.Greens, edgecolor='none',s=50)

# Set chart title and label axis
plt.title("Scatter Plot Square Numbers", fontsize=24)
plt.xlabel("value", fontsize=18)
plt.ylabel("Square of x", fontsize=18)

# Set size of tick labels
plt.tick_params(axis='both',which='major', labelsize=14)

plt.show()
```

```python
# Cubes

import matplotlib.pyplot as plt

xvalues = list(range(1,100))
yvalues = [x**3 for x in xvalues]
#print(yvalues)

plt.scatter(xvalues,yvalues,s=30)

#Name the plot
plt.title('Number vs Cube Plot', fontsize=25)
plt.xlabel('x', fontsize=14)
plt.ylabel('Cube(x)',fontsize=14)

#Set size of tick labels
plt.tick_params(axis='both', which='major', labelsize=14)

#Save the Image
#plt.show()
plt.savefig('images/cube.png',bbox_inches='tight')
```

```python
# Cube with color map
# Cubes

import matplotlib.pyplot as plt

xvalues = list(range(1,200))
yvalues = [x**3 for x in xvalues]
#print(yvalues)

plt.scatter(xvalues,yvalues,c=yvalues,cmap=plt.cm.Blues,edgecolor='none',s=50)

#Name the plot
plt.title('Number vs Cube Plot', fontsize=25)
plt.xlabel('x', fontsize=14)
plt.ylabel('Cube(x)',fontsize=14)

#Set size of tick labels
plt.tick_params(axis='both', which='major', labelsize=14)

#Save the Image
#plt.show()
plt.savefig('images/cube_cmap.png',bbox_inches='tight')
```
